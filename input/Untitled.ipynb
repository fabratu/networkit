```python
import networkit as nk

```

```python
G = nk.readGraph("network.dat")
attr = G.attachNodeAttribute("myAttr", str)
for u in G.iterNodes():
    attr[u] = str(u)

nk.plot.nodeAttributes(G,attr)

nk.plot.degreeDistribution(G, color = "green")
```

```python
G2 = nk.generators.ErdosRenyiGenerator(1000,0.002).generate()
nk.plot.connectedComponentsSizes(G2,True)
```
